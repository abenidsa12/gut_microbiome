```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.manifold import MDS
```

```python
path = os.path.abspath('D:/Kaiserslautern/Documents/Course content/Thesis/Excel sheets')
data = os.path.join(path, 'main-table17.csv')
df = pd.read_csv(data)
df.head()
```

```python
group_1 = ['N.vespillo_b', 'N.vsps_j', 'Yab_fpj_1', 'Yab_fpj_2', 'Yaj_fpb_1']
group_1_df = df[group_1]

colors = ['red', 'blue', 'green', 'orange', 'purple'] 

mds = MDS(n_components=2, metric=False, random_state=42)
transformed_data = mds.fit_transform(group_1_df.T)
plt.scatter(transformed_data[:, 0], transformed_data[:, 1])

for i, column_name in enumerate(group_1_df.T.index):
    plt.scatter(transformed_data[i, 0], transformed_data[i, 1], color=colors[i], label=column_name)

plt.xlabel('NMDS 1')
plt.ylabel('NMDS 2')
plt.title('NMDS Plot')
plt.legend(loc='lower right')

plt.show()
```
