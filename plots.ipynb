```python
selected_columns = [col for col in df.columns if col.startswith('Ya') or ('Ya' in col and 'fp' in col)]
fed_columns = [col for col in df.columns if col.startswith('Ya') and ('Ya' in col and 'fp' in col and 'fed' in col)]
non_fed_columns = [col for col in df.columns if col.startswith('Ya') and ('Ya' in col and 'fp' in col and 'fed' not in col)]
fed_df = df[fed_columns]
non_fed_df = df[non_fed_columns]
selected_df = df[selected_columns]
selected_df
```

```python
from sklearn.metrics import pairwise_distances
# combined_df = selected_df
combined_df = pd.concat([fed_df, non_fed_df])
combined_df = combined_df[~(combined_df == 0).all(axis=1)]
# print(combined_df)
combined_df.replace([np.inf, -np.inf], np.nan, inplace=True)
combined_df = combined_df.fillna(0)

distance_matrix = pairwise_distances(combined_df, metric="braycurtis")
np.fill_diagonal(distance_matrix, 0)
np.nan_to_num(distance_matrix, copy=False)

mds = MDS(n_components=2, dissimilarity="precomputed", random_state=42)
results = mds.fit_transform(distance_matrix)

plt.figure(figsize=(10, 8))

plt.scatter(results[:len(fed_df), 0], results[:len(fed_df), 1], color='red', label='Fed')
plt.scatter(results[len(non_fed_df):, 0], results[len(non_fed_df):, 1], color='blue', label='Non-Fed')

plt.legend()
plt.xlabel('PCoA1')
plt.ylabel('PCoA2')
plt.title('PCoA of Fed vs. Non-Fed')
plt.grid(True)
plt.show()
```
